### Read Image from Image URL

```python
url = "https://images.pexels.com/photos/3184398/pexels-photo-3184398.jpeg?auto=compress&cs=tinysrgb&h=750&w=1260"
```

```python
response = requests.get(url)
image = np.array(PIL.Image.open(BytesIO(response.content)))
```

### Face Encoding

It returns measures from the image which are used for comparison. If the person is same than these measures would be close.

```python
face_encoding = face_recognition.face_encodings(image)
```

### Displaying first face encoding only

```python
if len(face_encoding) == 0:
  print("No Face were Found")
else: 
  first_face_encoding = face_encoding[0]

  print(first_face_encoding)
```
